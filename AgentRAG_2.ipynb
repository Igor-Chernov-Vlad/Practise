````python


def final_model(question,context): #call llm to final answer

  prompt = """Using this context:{context}
  Answer the question:{question}
 """.format(question = question,context = context)
  messages = [
    {"role": "user", "content": prompt}
  ]
  print(0)
  text = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True,
      enable_thinking=False
  )
  model_inputs = tokenizer([text], return_tensors="pt").to(model.device)


  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=300
  )
  output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()

  try:

      index = len(output_ids) - output_ids[::-1].index(151668)
  except ValueError:
      index = 0

  thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
  content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")

  return content
````

````python
def generate_subQA(question,n): #split user questions on n subqestions
  prompt = "Split the question {question} on {n} subqestions. Use this template for answer 'qa:subqestion'".format(question = question,n = n)
  messages = [
      {"role": "user", "content": prompt}
  ]
  text = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True,
      enable_thinking=False
  )
  model_inputs = tokenizer([text], return_tensors="pt").to(model.device)


  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=600
  )
  output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()


  try:

      index = len(output_ids) - output_ids[::-1].index(151668)
  except ValueError:
      index = 0
  res = []
  thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
  content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")
  content = content.split('\n')
  for cont in content:
    if ("?" in cont) == True:
      res.append(cont)


  return res
````

````python
def Is_enough(question,context): # Call LLM to determine if there is enough information in the chunk

  prompt = """Is this context:{context} sufficient to fully answer the question:{question}?
  Answer Template: If sufficient, the answer is <yes>, if not sufficient, the answer is <no>.
 """.format(question = question,context = context)
  messages = [
    {"role": "user", "content": prompt}
  ]

  text = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True,
      enable_thinking=False
  )
  model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=300
  )
  output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()


  try:

      index = len(output_ids) - output_ids[::-1].index(151668)
  except ValueError:
      index = 0

  thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
  content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")

  res = False

  if ('yes' in  content ) == True or ('Yes' in  content )  == True or ('YES' in  content )  == True:
    res = True
  else:
    res = False

  return res
````

````python
def generate_follow_qa(context): # call LLM to generate follow questions

  prompt = """ Generate follow up question using this context:{context}
 """.format(context = context)
  messages = [
    {"role": "user", "content": prompt}
  ]

  text = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True,
      enable_thinking=False
  )
  model_inputs = tokenizer([text], return_tensors="pt").to(model.device)


  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=300
  )
  output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()

  try:

      index = len(output_ids) - output_ids[::-1].index(151668)
  except ValueError:
      index = 0

  thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
  content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")

  return content
````

````python
def mapreduce_find_context(question,subqestion,history,num_followqa): #Use MapReduce to summarize best information from each found chunk


  print(4)
  conts = vector_search(subqestion,3) # find most n releveant docs
  if len(history) != 0:
    for h in history:
      conts.append(h)
  addit_context = ''
  for count in range(num_followqa + 1):
    if len(addit_context) != 0:
      conts.append(addit_context)
      addit_context = ''


    d = {'context':conts}
    data = pd.DataFrame(data = d)
    doccontext = data['context'].iloc[0:]
    context = [Document(page_content=t) for t in doccontext]

    addmr =""" ```
    {context}
    ```
    """
    addr= """

    ```
    {best information found}
    ```
    <//>
    """



    map_template = """A user asked a question:'{question}'. The following is a set of contexts that the search found to answer this question, they are delimeted by ``` .
    Summarize the best information in this set according to relevance to the user's question and сreate a list of the best information found.


    """.format(question = question)
    map_template = map_template + addmr


    llm = HuggingFacePipeline(pipeline=pipe)
    map_prompt = PromptTemplate.from_template(map_template)
    map_chain = LLMChain(llm=llm,prompt=map_prompt)


    reduce_template = """The following is a set of best information found,which are delimeted by ``` .
    Look at the set of best information found and summarize the information found according to relevance to the user's question:'{question}'.



    """.format(question = question)
    reduce_template = reduce_template +  addr


    reduce_prompt = PromptTemplate(template=reduce_template,
                                                                    input_variables=["best information found"]

    )
    reduce_chain=LLMChain(llm=llm,prompt=reduce_prompt)

    combine_documents_chain = StuffDocumentsChain(
      llm_chain=reduce_chain, document_variable_name="best information found"
  )
    reduce_documents_chain = ReduceDocumentsChain(
        combine_documents_chain=combine_documents_chain,
        collapse_documents_chain=combine_documents_chain,
        token_max=40000)
    map_reduce_chain = MapReduceDocumentsChain(
        llm_chain=map_chain,
        reduce_documents_chain=reduce_documents_chain,
        document_variable_name="context",
        return_intermediate_steps=False)

    output = map_reduce_chain.run(context)
    print('13',output )
    resl = output.split('<//>')
    res = resl[1]

    isEnoug = Is_enough(question,res)
    if isEnoug == True:
      final_context.append(res)
      break
    else:
      if num_followqa == (count - 1 ):
        final_context.append(res)
        break
      else:
        addit_context = generate_follow_qa(res)

````

````python
from transformers import AutoModelForCausalLM, AutoTokenizer
from deepeval.models.base_model import DeepEvalBaseLLM
class Qwen_Qwen3_0_6B(DeepEvalBaseLLM):
    def __init__(self):


        model_4bit = AutoModelForCausalLM.from_pretrained(
            "Qwen/Qwen3-0.6B",
            device_map="auto",

        )
        tokenizer = AutoTokenizer.from_pretrained(
            "Qwen/Qwen3-0.6B"
        )

        self.model = model_4bit
        self.tokenizer = tokenizer

    def load_model(self):
        return self.model

    def generate(self, prompt: str) -> str:
        model = self.load_model()

        pipeline = transformers.pipeline(
            "text-generation",
            model=model,
            tokenizer=self.tokenizer,
            use_cache=True,
            device_map="auto",
            max_length=2500,
            do_sample=True,
            top_k=5,
            num_return_sequences=1,
            eos_token_id=self.tokenizer.eos_token_id,
            pad_token_id=self.tokenizer.eos_token_id,
        )

        return pipeline(prompt)

    async def a_generate(self, prompt: str) -> str:
        return self.generate(prompt)

    def get_model_name(self):
        return "Qwen3-0.6B"
````

````python
from deepeval.test_case import LLMTestCase

def evalute_contextual_precision(input,actual_output,expected_output,retrieval_context):
  input = input.tolist()
  actual_output = actual_output.tolist()
  expected_output = expected_output.tolist()
  retrieval_context = retrieval_context.tolist()
  test_case = LLMTestCase(
      input=input,
      actual_output=actual_output,
      expected_output=expected_output ,
      retrieval_context=[retrieval_context])
  contextual_precision.measure(test_case)
  return contextual_precision.score , contextual_precision.reason
````

````python

def evalute_contextual_recall(input,actual_output,expected_output,retrieval_context):
  input = input.tolist()
  actual_output = actual_output.tolist()
  expected_output = expected_output.tolist()
  retrieval_context = retrieval_context.tolist()
  test_case = LLMTestCase(
      input=input,
      actual_output=actual_output,
      expected_output=expected_output ,
      retrieval_context=[retrieval_context])
  contextual_recall.measure(test_case)
  return contextual_precision.score , contextual_precision.reason
````
